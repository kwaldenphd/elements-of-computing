# Area Plots

An area plot is essentially a "filled" line plot.

```python
import pandas as pd, numpy as np # import statements
df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"]) # create data
df.plot.area() # show plot
```
